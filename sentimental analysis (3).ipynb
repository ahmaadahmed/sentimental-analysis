```python
# Define the number of records to select for each category
records_per_category = 4000

# Create a function to sample records for each category
def sample_records(category):
    return df[df['target'] == category].sample(records_per_category)

# List to store sampled DataFrames for each category
sampled_dfs = []

# Loop through each category (1, 2, 3, 4, 0)
for category in range(0, 5):
    sampled_dfs.append(sample_records(category))

# Concatenate the sampled DataFrames into a final DataFrame
result_df = pd.concat(sampled_dfs)

# Display the resulting DataFrame
print(result_df)
```

```python
# Assuming your CSV file has columns 'text' and 'label'
data = list(zip(ar_df['text'], ar_df['target']))
```

```python
# Split the dataset into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# Define a custom dataset class
class SentimentDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=128):
        self.data = data
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text, label = self.data[idx]
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'label': torch.tensor(int(label == 'positive'))  # Assuming binary classification
        }
```

```python
# Load the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertForSequenceClassification.from_pretrained('bert-base-multilingual-cased', num_labels=2)  # 2 for binary classification
```

```python
# Create datasets and dataloaders
train_dataset = SentimentDataset(train_data, tokenizer)
test_dataset = SentimentDataset(test_data, tokenizer)
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=8, shuffle=False)
```

```python
# Use GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
# Define training parameters
optimizer = AdamW(model.parameters(), lr=2e-5)
num_epochs = 10
```

```python
# Training loop
for epoch in range(num_epochs):
    model.train()
    for batch in tqdm(train_dataloader, desc=f'Epoch {epoch + 1}/{num_epochs}'):
        inputs = {key: val.to(device) for key, val in batch.items() if key != 'label'}
        labels = batch['label'].to(device)
        
        optimizer.zero_grad()
        outputs = model(**inputs, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
```

```python
# Evaluation
model.eval()
predictions, true_labels = [], []
for batch in tqdm(test_dataloader, desc='Evaluating'):
    inputs = {key: val.to(device) for key, val in batch.items() if key != 'label'}
    labels = batch['label'].to(device)
    
    with torch.no_grad():
        outputs = model(**inputs)
    
    logits = outputs.logits
    predictions.extend(torch.argmax(logits, dim=1).cpu().numpy())
    true_labels.extend(labels.cpu().numpy())
```

```python
# Calculate accuracy
accuracy = accuracy_score(true_labels, predictions)
print(f'Accuracy: {accuracy:.4f}')
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
```

```python
# Load your trained model (replace 'path/to/your/model' with the actual path)
model_path = r"E:\sentimental analysis code\content1\model"
model = BertForSequenceClassification.from_pretrained(model_path)
tokenizer = BertTokenizer.from_pretrained(model_path)
```

```python
def convert_sentiment_label(label):
    label_names = {
        0: 'فن',
        1: 'قضاء',
        2: 'اقتصاد',
        3: 'سياسة',
        4: 'رياضة'
    }

    return label_names[label]
```

```python
# Sentence to evaluate
sentence = "تعرف عليها الجمهور المغربي في الأغاني الطربية"

# Tokenize and prepare the input for the model
inputs = tokenizer(sentence, return_tensors='pt')

# Forward pass through the model
with torch.no_grad():
    outputs = model(**inputs)

# Get the predicted label
predicted_label = torch.argmax(outputs.logits).item()

sentiment_name = convert_sentiment_label(predicted_label)
print(sentiment_name)
```
